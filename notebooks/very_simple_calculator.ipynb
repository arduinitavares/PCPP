# Scenario
You need a calculator. A very simple and very specific calculator. Look at the picture - it contains two fields that the user can use to enter arguments, a radio button to select the operation to perform, and a button initiating the evaluation:

# Calculator - reference


We expect the calculator to behave in the following way:

* if both fields contain valid (integer or float) numbers, clicking the Evaluate button should display an info window showing the evaluation's result;
* if any of the fields contains invalid data (e.g., a string, or a field is empty), clicking the Evaluate button should present an error window describing the problem, and the focus should be moved to the field causing the problem.
Don't forget to protect your code from dividing by zero, and use the grid manager to compose the window interior.



```python
import tkinter as tk
from tkinter import messagebox


def evaluate():
    try:
        left_operand = float(text_left.get())
        right_operand = float(text_right.get())
        if operation.get() == "Add":
            result = left_operand + right_operand
        elif operation.get() == "Subtract":
            result = left_operand - right_operand
        elif operation.get() == "Multiply":
            result = left_operand * right_operand
        elif operation.get() == "Divide":
            result = left_operand / right_operand
        messagebox.showinfo("Result", f"The result is: {result}")
    except ValueError:
        messagebox.showerror("Error", "Invalid input")
    except ZeroDivisionError:
        messagebox.showerror("Error", "Cannot divide by zero")
    except Exception as e:
        messagebox.showerror("Error", f"An error occurred: {e}")
    finally:
        text_left.set("")
        text_right.set("")
        operation.set("Add")
        entry_left.focus()
        entry_right.focus()

# Create the main window
window = tk.Tk()

# Create a StringVar object, which is a special type of variable for use with tkinter widgets
text_left = tk.StringVar()
text_right = tk.StringVar()

# Create an Entry widget to display the text
entry_left = tk.Entry(window, textvariable=text_left)
entry_right = tk.Entry(window, textvariable=text_right)


# Add the Entry widget to the window
entry_left.grid(row=2, column=0)
entry_right.grid(row=2, column=2)

# Create button "Evaluate"
button_evaluate = tk.Button(window, text="Evaluate", command=evaluate)

# Add the button to the window
button_evaluate.grid(row=4, column=1)

# Create a StringVar object for the radio buttons
operation = tk.StringVar()
radio_1 = tk.Radiobutton(window, text="Add", variable=operation, value='Add')
radio_2 = tk.Radiobutton(window, text="Subtract", variable=operation, value='Subtract')
radio_3 = tk.Radiobutton(window, text="Multiply", variable=operation, value='Multiply')
radio_4 = tk.Radiobutton(window, text="Divide", variable=operation, value='Divide')

radio_1.grid(row=0, column=1)
radio_2.grid(row=1, column=1)
radio_3.grid(row=2, column=1)
radio_4.grid(row=3, column=1)


# Initialize a default operation
operation.set("Add")

window.mainloop()
```
